### Cleaning Data in Country columns

```python
def ChangeCountry(x):
    if 'North ' in x:
        x = x.replace('North ','') 
    elif 'The ' in x:
        x = x.replace('The ','') 
    elif ', The' in x:
        x = x.replace(', The','') 
    return x

df['country'] = df['country'].apply(ChangeCountry)
```

### Normality Test

```python
import statsmodels.api as sm
import pylab
from scipy.stats import anderson,normaltest

def uji_normalitas(data):
    dfNorm = pd.DataFrame({
        'Statistic' : anderson(data)[0],
        'Critical Value' : anderson(data)[1],
        'Significance Level (%)' : anderson(data)[2]
    })
    listmsg= []
    for i in range(len(dfNorm['Statistic'])):
        if dfNorm['Statistic'].values[i] > dfNorm['Critical Value'].values[i]:
            msg =('Tolak H0, Sample Not Has Normal Distribution')
        else:
            msg =('Terima H0, Sample Has Normal Distribution')
        listmsg.append(msg)
    dfNorm['Conclusion'] = listmsg
    
    fig,ax = plt.subplots(ncols=2,nrows=1,figsize=(10,4))
    sns.distplot(data,color = 'red',ax=ax[0])
    ax[0].set_xlabel("Frequency")
    ax[0].set_title('Distribution Plot')
    
    sm.qqplot(data, line='45',ax=ax[1])
    plt.title('QQ Plot Testing')
    pylab.show()
    
    plt.tight_layout
    
    return dfNorm

uji_normalitas(df_last[df_last['climate']=='nontropic']['confirmed'])
```

### Regression Model
> Country = US, Date = 2020-03-20 until 2020-08-10

```python
start_date = '2020-03-20'
end_date = '2020-08-10'
US = df_new[df_new['country']=='US']
date = (US['date'] >= start_date) & (US['date'] <= end_date)
US = US.loc[date]
US
```

```python
# Feature Selection
Xreg = US[['confirmed', 'recovered']]
yreg = US['deaths']
```

```python
# Train Test Split
Xreg_train, Xreg_test, yreg_train, yreg_test = train_test_split(Xreg,yreg,test_size=0.3,random_state=101)
```

```python
# Check the best model to predict this target

def regression_model(n_cv,x,y):
    score = []
    model_name = ['Linear Regression','DecisionTreeRegressor','RandomForestRegressor','GradientBoostingRegressor',
                  'XGBRegressor']
    models = [LinearRegression,DecisionTreeRegressor,RandomForestRegressor,GradientBoostingRegressor,XGBRegressor]
    
    for item in models:
        mae = []
        rmse = []
        r2score = []
        kf = KFold(n_splits=n_cv,shuffle=True,random_state=101)
        for train_index , test_index in kf.split(x,y):
            X_train,X_test,y_train,y_test = x.iloc[train_index],x.iloc[test_index],y.iloc[train_index],y.iloc[test_index]
            model = item()
            model.fit(X_train,y_train)
            prediction = model.predict(X_test)
            mae.append(round(mean_absolute_error(y_test,prediction),1))
            rmse.append(round(np.sqrt(mean_squared_error(y_test,prediction)),1))
            r2score.append(round(metrics.r2_score(y_test,prediction),1))            

        temp = [
            sum(mae)/n_cv,
            sum(rmse)/n_cv,
            sum(r2score)/n_cv
        ]
        score.append(temp)
    return pd.DataFrame(score,columns=['MAE','RMSE','R2Score'],index=model_name).sort_values(by='R2Score',ascending=False)
```

```python
regression_model(5,Xreg,yreg)
```

> Based on validation data using KFold, RandomForestRegressor have a best performance because it produce little MAE and RMSE Score and have big score for R2Score.



```python
# Modelling menggunakan RandomForestRegressor
RFreg = RandomForestRegressor(random_state=101)
RFreg.fit(Xreg_train, yreg_train)
prediction = RFreg.predict(Xreg_test)
print('Evaluation Result: \n')
print(f'MAE : {metrics.mean_absolute_error(yreg_test,prediction)}')
print(f'MSE : {metrics.mean_squared_error(yreg_test,prediction)}')
print(f'RMSE : {np.sqrt(metrics.mean_squared_error(yreg_test,prediction))}')
print(f'R2_Score : {(metrics.r2_score(yreg_test,prediction))*100}%')
```

> If we see from the evaluation results only, this model is well enough

### Clustering Model
> I used KMeans Clustering to divide this features into four different cluster based on number of confirmed, recovered and deaths cases. 

```python
Z=df_last.loc[:,['confirmed','recovered','deaths']].values
pd.DataFrame(Z,columns=['confirmed','recovered','deaths'])
```

```python
#Scaling Data
scaler = StandardScaler()
scaler.fit(Z)
scaledZ = scaler.transform(Z)
#Check Value in dataset after scaling
scaledZ
```

```python
# Do decompositioning by PCA
from sklearn.decomposition import PCA
pca_2 = PCA(2)
pca_2_result = pca_2.fit_transform(scaledZ)

pca_2.explained_variance_ratio_.cumsum()
```

```python
# Modelling with KMeans Clustering
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=5,
               init='k-means++',
               n_init=15,
               max_iter=500,
               random_state=101)
kmeans_labels = kmeans.fit_predict(scaledZ)

centroids = kmeans.cluster_centers_
centroids_pca = pca_2.transform(centroids)
```

```python
fig,ax = plt.subplots(1,2,figsize=(12,6))
# Data Original
ax[0].scatter(x=pca_2_result[:, 0], y=pca_2_result[:, 1],lw=0.1)
ax[0].set_title('Original Data')
ax[0].set_xlabel('Principal Component 1')
ax[0].set_ylabel('Principal Component 2')
# Data Menggunakan KMeans
ax[1].scatter(x=pca_2_result[:, 0], y=pca_2_result[:, 1],lw=0.1,c=kmeans_labels,cmap="RdYlGn")
ax[1].set_title('Clustering Results using K-Means')
ax[1].scatter(centroids_pca[:,0],centroids_pca[:,1],c='red',marker='*',s=200)
ax[1].set_xlabel('Principal Component 1')
ax[1].set_ylabel('Principal Component 2')
plt.tight_layout()
plt.show()
```

```python
# Apply result cluster to dataset
clustering = df_last.copy()
clustering['cluster'] = kmeans_labels
clustering
```
